```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
bottom_up_data = np.loadtxt("result31.txt", dtype=int)
bottom_up_costs = bottom_up_data[:, 0]
bottom_up_x = np.arange(0, 350001, 10000)
bottom_up_x[-1] = 346025
```

```python
top_down_data = np.loadtxt("result32.txt", dtype=int)
top_down_costs = top_down_data[:, 0]
top_down_x = np.arange(0, 150000 + 1, 10000)
top_down_x[-1] = 143535
```

```python
fig, ax = plt.subplots()
ax.plot(bottom_up_x, bottom_up_costs, label="Parcours ascendant")
ax.plot(top_down_x, top_down_costs, label="Parcours descendant")
ax.axhline(y=820800, color="black", linestyle="dotted")
ax.set_xlabel("Itrations")
ax.set_ylabel("Cot des oprations")
ax.legend()
fig.savefig("progres_cout_operations.pdf", bbox_inches="tight")
```
